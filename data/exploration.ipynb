```python
import pandas as pd

df = pd.read_csv("real-estate-raw-6a3e627c94437783777235.csv")
print(f"Shape: {df.shape}")
df.head()
```

```python
print("=== SHAPE ===")
print(df.shape)

print("\n=== COLUMN TYPES ===")
print(df.dtypes)

print("\n=== MISSING VALUES ===")
print(df.isnull().sum())

print("\n=== DUPLICATES ===")
print(f"Duplicate listing_ids: {df['listing_id'].duplicated().sum()}")
```

```python
print("\n=== PRICE SAMPLES ===")
print(df['price'].value_counts().head(10))

print("\n=== PROPERTY TYPE VALUES ===")
print(df['property_type'].value_counts())

print("\n=== PARKING VALUES ===")
print(df['parking'].value_counts())

print("\n=== HEATING TYPE VALUES ===")
print(df['heating_type'].value_counts())

print("\n=== DATE SAMPLES ===")
print(df['listing_date'].value_counts().head(10))

print("\n=== PRICE STATS ===")
df['price_clean'] = pd.to_numeric(df['price'], errors='coerce')
print(df['price_clean'].describe())

print("\n=== SURFACE STATS ===")
print(df['surface_m2'].describe())
```
